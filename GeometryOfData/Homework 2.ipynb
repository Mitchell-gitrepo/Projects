<p>d) <br>
All Christoffel symbols were equal to 0 except for three of them. As such, to save space in this document I will only be deriving the Christoffel symbols for values of $i,j,k$ where the value of their Christoffel symbols is greater than 0.<br><br>
$k=1,i=1,j=2$: $\space \Gamma_{1,2}^1 \space = \space \frac{1}{2}(g^{1,1}(\frac{\partial g_{2,1}}{\partial x^1}+\frac{\partial g_{1,1}}{\partial x^2}-\frac{\partial g_{1,2}}{\partial x^1})+g^{1,2}(\frac{\partial g_{2,2}}{\partial x^1}+\frac{\partial g_{1,2}}{\partial x^2}-\frac{\partial g_{1,2}}{\partial x^2})) \space = \space \frac{1}{2}(\frac{1}{cos^2\phi}(0+\frac{\partial (cos^2\phi)}{\partial \phi}-0)+0(0+0-0)) \space = \space \frac{1}{2}(\frac{1}{cos^2\phi}(-2sin\phi cos\phi)) \space = \space -\frac{sin\phi}{cos\phi}$<br><br>
    
$k=1,i=2,j=1$: $\space \Gamma_{2,1}^1 \space = \space \frac{1}{2}(g^{1,1}(\frac{\partial g_{1,1}}{\partial x^2}+\frac{\partial g_{2,1}}{\partial x^1}-\frac{\partial g_{2,1}}{\partial x^1})+g^{1,2}(\frac{\partial g_{1,2}}{\partial x^2}+\frac{\partial g_{2,2}}{\partial x^1}-\frac{\partial g_{2,1}}{\partial x^2})) \space = \space \frac{1}{2}(\frac{1}{cos^2\phi}(\frac{\partial (cos^2\phi)}{\partial \phi}+0-0)+0(0+0-0)) \space = \space \frac{1}{2}(\frac{1}{cos^2\phi}(-2sin\phi cos\phi)) \space = \space -\frac{sin\phi}{cos\phi}$<br><br>
    
$k=2,i=1,j=1$: $\space \Gamma_{1,1}^2 \space = \space \frac{1}{2}(g^{2,1}(\frac{\partial g_{1,1}}{\partial x^1}+\frac{\partial g_{1,1}}{\partial x^1}-\frac{\partial g_{1,1}}{\partial x^1})+g^{2,2}(\frac{\partial g_{1,2}}{\partial x^1}+\frac{\partial g_{1,2}}{\partial x^1}-\frac{\partial g_{1,1}}{\partial x^2})) \space = \space \frac{1}{2}(0(0+0-0)+1(0+0-\frac{\partial (cos^2\phi)}{\partial \phi})) \space = \space \frac{1}{2}((-2sin\phi cos\phi)) \space = \space -sin\phi cos\phi$

<p>e) <br>
$\gamma(t)=s(t,0)$ is a geodesic if it's partial second derivative satisfies the below equation:<br><br>
$\frac{\partial^2 \gamma(t)^k}{\partial t^2} \space = \space \sum\limits_{i,j=1}^{d} \Gamma_{i,j}^k (\gamma(t)) \frac{\partial \gamma^i}{\partial t}\frac{\partial \gamma^j}{\partial t}$<br><br>
    
Note that $\frac{\partial^2 \gamma(t)^k}{\partial t^2} \space = \space [\frac{\partial^2 (t)}{\partial t^2}, \frac{0}{\partial t^2}]^T=[0,0]^T$<br><br>
    
Then let us consider the case of $k=1$. The only Christoffel symbols that contribute in this case are for $k=1,i=1,j=2$ and $k=1,i=2,j=1$ as shown above. Then: <br><br>
    
$\frac{\partial^2 \gamma(t)^1}{\partial t^2} \space = \space \Gamma_{1,2}^1 (\gamma(t)) \frac{\partial \gamma^1}{\partial t}\frac{\partial \gamma^2}{\partial t}+\Gamma_{2,1}^1 (\gamma(t)) \frac{\partial \gamma^2}{\partial t}\frac{\partial \gamma^1}{\partial t} \space = \space \frac{-sin\phi}{cos\phi} (\gamma(t)) \frac{\partial (t)}{\partial t}\frac{\partial 0}{\partial t}+\frac{-sin\phi}{cos\phi} (\gamma(t)) \frac{\partial (0)}{\partial t}\frac{\partial (t)}{\partial t} \space = \space 0$ since $\phi=0$<br><br>

For the case of $k_=2$ only one Christoffel symbol contributes, which in this case is for $k=2, i=1, j=1$. Then:
    
$\frac{\partial^2 \gamma(t)^2}{\partial t^2} \space = \space \Gamma_{1,1}^2 (\gamma(t)) \frac{\partial \gamma^1}{\partial t}\frac{\partial \gamma^1}{\partial t} \space = \space -sin\phi cos\theta (\gamma(t)) \frac{\partial (t)}{\partial t}\frac{\partial (t)}{\partial t} \space = \space 0$ since $\phi=0$<br><br>
    
So then it has been shown that the equator circle is a geodesic as it satisfies the above differential equation.
</p>

<h3>Methods</h3>

```python
import numpy as np
import math

def preprocess(shapeArr):
    newShapeArr = []
    for shape in shapeArr:
        mean = np.mean(shape, axis=0)
        shape = shape - mean                            #remove translation
        froNorm = np.linalg.norm(shape)
        shape = shape / froNorm                          #remove scaling
        newShapeArr.append(np.transpose(shape))
    return newShapeArr


def opa(targetShape, shape):
    multShapes = np.matmul(targetShape, np.transpose(shape))
    u, s, v = np.linalg.svd(multShapes, full_matrices=False)
    s = s * [[1,0],[0,1]]
    rotate = np.matmul(u, np.transpose(v))
    reflect_direction = 0
    
    while np.linalg.det(rotate) < 0:                     #We only want rotation - no reflection!
        reflect_yaxis = [[-1,0],[0,1]]
        reflect_xaxis = [[1,0],[0,-1]]
        if (reflect_direction % 2) == 0:                  #This is to remove any reflection that may occur
            rotate = np.matmul(reflect_yaxis,rotate)
        else:
            rotate = np.matmul(reflect_xaxis,rotate)
        reflect_direction = reflect_direction+1
    return np.matmul(rotate,shape)


def rotateShapeArr(targetShape, shapeArr):
    newShapeArr = []
    count = 1
    for i in range(0,len(shapeArr)):
        shape = shapeArr[i]
        newShapeArr.append(opa(targetShape,shape))
        count = count + 1
    return newShapeArr


def expMap(p, tanVector):
    theta = np.linalg.norm(tanVector)
    if theta ==0:                                           #if tanVector has norm of 0, then stay at p!
        return p
    else:
        q = p * math.cos(theta) + tanVector/theta * math.sin(theta)
    return q


def logMap(p, q):
    m = p.shape[0]
    n = p.shape[1]
    p = p.flatten()
    q = q.flatten()
    try:
        theta = math.acos(np.dot(p,q))                      #argument for math.acos cannot be greater than 1
    except:
        theta = math.acos(round(np.dot(p,q)))               #argument for math.acos cannot be greater than 1
    if theta == 0:                                          #if theta is zero the tangent vector has all 0 entries.
        tanVector = np.zeros((m,n))
    else:
        tanVector = theta * ((q-p*math.cos(theta))/math.sin(theta))
        tanVector = tanVector.reshape(m, n)
    return tanVector


def frechetMean(shapeArr, epsilon):
    mean = shapeArr[0]
    size = mean.shape
    delta_mean =np.zeros((size[0],size[1])) + epsilon + 1      #initializing delta mean value
    while np.linalg.norm(delta_mean) > epsilon:
        sum = np.zeros((size[0],size[1]))
        for shape in shapeArr:
            shape = opa(mean, shape)
            sum = sum + logMap(mean, shape)
        delta_mean = sum / len(shapeArr)
        mean = expMap(mean, delta_mean)
    return mean


def pga(frechetMean, shapeArr):
    m = shapeArr[0].shape[0]                                   #number of row
    n = shapeArr[0].shape[1]                                   #number of columns
    S = np.zeros((m*n,m*n))
    for shape in shapeArr:
        shape = opa(frechetMean, shape)
        u = logMap(frechetMean, shape)
        u = u.flatten()
        u = u.reshape(m*n, 1)
        S = S + np.matmul(u,np.transpose(u))
    S = S * (1/(len(shapeArr)))
    eigVals, eigVectors = np.linalg.eigh(S)
    return eigVals, eigVectors
```

<br>Generating the 100 triangles

```python
import numpy as np
import math

#Generate 100 triangles
triangleData = []
mean = 1
sd = 0.5
noise_mean = 0
noise_sd = 0.1
for i in range(0,100):
    s = np.random.normal(mean,sd)
    triangle = [[-1+np.random.normal(noise_mean,noise_sd),0+np.random.normal(noise_mean,noise_sd)],
                [1+np.random.normal(noise_mean,noise_sd),0+np.random.normal(noise_mean,noise_sd)],
                [0+np.random.normal(noise_mean,noise_sd),s+np.random.normal(noise_mean,noise_sd)]]
    triangleData.append(triangle)

print("done")
```

<br><br>Plotting a geodesic between two shapes - The result is a plot that shows how the shapes deform as we travel along the geodesic connecting the two

```python
#Plot a geodesic between two points
from matplotlib import pyplot as plt
newTriData = preprocess(triangleData)
triangleFrechet = frechetMean(newTriData, .0001)
triPreshapeAligned = rotateShapeArr(triangleFrechet, newTriData)
object1 = triPreshapeAligned[0]
object2 = triPreshapeAligned[1]
geoVector = logMap(object1, object2)
geo_Interval = [0,.25,.5,.75,1]
legend = ['0%','25%','50%','75%','100%']

    
#Deformation
plt.title("Geodesic Transformation") 
plt.xlabel("X-Axis") 
plt.ylabel("Y-Axis")
geo_Shapes = []

for g in geo_Interval: 
    geo_Shapes.append(expMap(object1, g * geoVector))

for i in range(0,len(geo_Shapes)):
    geoShape = geo_Shapes[i]
    geoShapeX = geoShape[0]
    geoShapeX = np.append(geoShapeX,geoShape[0][0])
    geoShapeY = geoShape[1]
    geoShapeY = np.append(geoShapeY,geoShape[1][0])
    plt.plot(geoShapeX,geoShapeY,label=legend[i])
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.show()
```

<br><br>Plot of raw object data, preshapes aligned to the frechet mean, and a plot of the frechet mean shape. I see a significant difference between the raw object data and the preshapes aligned to the mean, in that the aligned preshape data seems to all "point" in the same direction as the mean.

```python
#Plot of raw object data
from matplotlib import pyplot as plt

plt.title("Original Triangle Data") 
plt.xlabel("X-Axis") 
plt.ylabel("Y-Axis")
for triangle in triangleData:
    triangleX = [pt[0] for pt in triangle]
    triangleX.append(triangle[0][0])
    triangleY = [pt[1] for pt in triangle]
    triangleY.append(triangle[0][1])
    plt.plot(triangleX,triangleY) 
plt.show()
```
