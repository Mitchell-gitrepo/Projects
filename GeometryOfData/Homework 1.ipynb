<p>Please find below my function to compute the stereographic projection of a 2-sphere onto &#8477;<sup>2</sup></p>

```python
def stereographic(x,y,z,radius):
    #Note that if the radius of the sphere is 1 (i.e. the north pole is (0,0,1) and south pole at (0,0,-1)), then 
    #the below expression simplifies to [x/(1-z),y/(1-z)]
    return [radius*x/(radius-z),radius*y/(radius-z)]
```

<h6>a) What is the image of a longitudinal line?</h6>
<br> <p>The first step is to define what a longitudinal line looks like in &#8477;<sup>3</sup>. I am going to define a longitudinal line by the angle that is formed by its stereographic projection in the x,y plane relative to the x-axis and y-axis, which I will call &theta;. So for instance, if &theta; = 0, then this is just the longitudinal line that passes through the points, (0,0,1), (1,0,0), (0,0,-1), and (-1,0,0). Similarly if &theta; = &pi;/2, then the corresponding longitudinal line would pass through the points (0,0,1), (0,1,0), (0,0,-1), and (0,-1,0).
<br>So since any point on the sphere is given by <em>x<sup>2</sup></em> + <em>y<sup>2</sup></em> + <em>z<sup>2</sup></em> = <em>r<sup>2</sup></em>, then given a z coordinate and &theta;, the x and y coorindates of a longitudinal line are given by cos&theta;<span>&#8730;(r<sup>2</sup>-z<sup>2</sup>)</span> and sin&theta;<span>&#8730;(r<sup>2</sup>-z<sup>2</sup>)</span>. Additionally if we define the z coordinate of the longitudinal line as sin(&phi;)r, where 0 &lt; &phi; &lt; 2&pi;, then we can define our longitudinal circle as having the coordinates (cos&theta;<span>&#8730;(1-z<sup>2</sup>)</span>, sin&theta;<span>&#8730;(1-z<sup>2</sup>)</span>, sin(&phi;)r)

<br>My code below attempts to calculate and graph the stereographic projection of these longitude lines using only the angle &theta; as an input.</p>

```python
import matplotlib.pyplot as plt
figure, axes = plt.subplots()
import math
pi = math.pi

sphere_radius = 1 #This corresponds to a sphere with north pole (0,0,1)

thetaList = [0, pi/4, pi/2, 3*pi/4, pi] #This is the angle that is being used to define longitudinal lines
rhoList = [i/100 * pi for i in range(0,50)] #In essance, I am discretizing the continuous longitudinal line into discrete points
for theta in thetaList:
    xStereo = [] #A list to store the x coordinates of the stereographic projection 
    yStereo = [] #A list to store the y coordinates of the stereographic projectsion 
    for rho in rhoList:
        
        zCoord = sphere_radius * math.sin(rho)
        xCoord = math.sqrt(sphere_radius**2-zCoord**2) * math.cos(theta)
        yCoord = math.sqrt(sphere_radius**2-zCoord**2) * math.sin(theta)
        stereographCoord = stereographic(xCoord, yCoord, zCoord, sphere_radius)
        
        #Since I only have
        xCoordMirror = math.sqrt(sphere_radius**2-zCoord**2) * math.cos(theta + pi)
        yCoordMirror = math.sqrt(sphere_radius**2-zCoord**2) * math.sin(theta + pi)
        stereographCoordMirror = stereographic(xCoordMirror, yCoordMirror, zCoord, sphere_radius)
        
        xStereo.append(stereographCoord[0])
        yStereo.append(stereographCoord[1])
        xStereo.append(stereographCoordMirror[0])
        yStereo.append(stereographCoordMirror[1])
    plt.plot(xStereo, yStereo, label=str("theta: "+ str(theta)))
plt.legend()
plt.show()
```

<h6>b) What is the image of a latitude line?</h6>
    <br><p>Note that given a constant <em>z</em>, the latitude line on a sphere is given by <em>x<sup>2</sup></em> + <em>y<sup>2</sup></em> + <em>z<sup>2</sup></em> = <em>r<sup>2</sup></em> and its stereographic projection is given by <em>x<sub>proj</sub><sup>2</sup></em> + <em>y<sub>proj</sub><sup>2</sup></em> = <em>r<sub>proj</sub><sup>2</sup></em> where <em>x<sub>proj</sub>, y<sub>proj</sub></em>, and <em>r<sub>proj</sub></em> are the x and y coordinates, and radius of the stereographically projected circle onto &#8477;<sup>2</sup> respectively. Note that using our stereographic projection defined earlier, we can show that <em>r<sub>proj</sub><sup>2</sup></em> is equal to ((<em>r<sub>0</sub>x<sub>0</sub></em>) / (<em>r<sub>0</sub>-z</em>))<sup>2</sup> + ((<em>r<sub>0</sub>y<sub>0</sub></em>) / (<em>r<sub>0</sub>-z</em>))<sup>2</sup> where <em>x<sub>0</sub>, y<sub>0</sub></em>, and <em>r<sub>0</sub></em> correspond to the x and y coordinates of the latitude line on the sphere and the radius of the sphere respectively. Simplifying this expression and substituting in <em>r<sub>0</sub><sup>2</sup></em> - <em>z<sup>2</sup></em> = <em>x<sub>0</sub><sup>2</sup></em> + <em>y<sub>0</sub><sup>2</sup></em> yields the equation for the projected latitude line on &#8477;<sup>2</sup> which is <em>x<sub>proj</sub><sup>2</sup></em> + <em>y<sub>proj</sub><sup>2</sup></em> = (<em>r<sub>0</sub><sup>2</sup></em> (<em>r<sub>0</sub></em> + <em>z</em>)) / (<em>r<sub>0</sub></em> - <em>z</em>)

<h6>c) What is the image of an arbitrary great circle?</h6>
<br> <p> This first step is to define what this circle would look like in &#8477;<sup>3</sup>. A great circle can be thought of as a longitudinal circle that passes through the points ((0,0,1), (0,1,0), (0,0,-1), (0,-1,0)) but rotated along the XY plane as well as the Z-axis. This makes it easier to define a great circle as we can think of it in terms of the angles it makes when the plane it lies on intersects with the XY plane and X-axis. We will define these two angles as follows:

<br>1) &theta; will denote the angle that is made by the intersection of the great circle and the XY plane. Then given &theta;, we can transform the z-coordinate for any point along the longitudinal line defined above by multiplying it by sin(&theta;). So for instance, if &theta;=[0, &pi;/4, &pi;/2], then the z-coordinate of the highest point of their great circles (assuming radius=1) would be [0, <span>>&#8730;2</span>/2, and 1] respectively.
    
<br>2) &rho; will denote the angle of the plane relative to the X-axis (i.e. if one were to draw a line from the highest point of the great circle down to its lowest point, and then take the orthogonal projection of this line onto &#8477;<sup>2</sup> then what angle would this orthogonally projected line make with the X-axis.</p>

```python
import matplotlib.pyplot as plt
figure, axes = plt.subplots()
import math
pi = math.pi

sphere_radius = 1 #This corresponds to a sphere with north pole (0,0,1)
thetaList = [0, pi/4, pi/2, 3*pi/4, pi]
rhoList = [0, pi/4, pi/2, 3*pi/4, pi]


#first start with the longitudinal line passing through ((0,0,1), (0,1,0), (0,0,-1), (0,-1,0))
#then transform X and Z components by setting them to cos(theta) and sin(theta) respectively
#then transform X and Y components by rotating the projection by rho

minCoord = 0#This is for determining the bounds of our scatterplot
maxCoord = 0#This is for determining the bounds of our scatterplot
for theta in thetaList:
    for rho in rhoList:
        x_orig = [] #A list to store the original x coordinates of the original great circle 
        y_orig = [] #A list to store the original y coordinates of the original great circle 
        xStereo = [] #A list to store the x coordinates of the stereographic projection 
        yStereo = [] #A list to store the y coordinates of the stereographic projectsion 
       
    
        for i in range(1,100): #starting at value of 1 and ending at i = 99, otherwise it's possible to get a divide by zero error
            #This first section creates a longitudinal line that passes through the points ((0,0,1), (0,1,0), (0,0,-1), (0,-1,0))
            #and then rotates by an angle of theta, which is the desired angle between the XY Plane and the great circle
            phi = (i/100) * (2 * pi)
            zCoord = sphere_radius * math.cos(phi) * math.sin(theta)
            yCoord = sphere_radius * math.sin(phi)
            if 25 < i < 75: #math.sqrt always returns positive numbers, so we have to specify when it should be negative,
                            #which should occur between pi/2 and 3pi/2
                xCoord = -math.sqrt(max(0,sphere_radius**2 - zCoord**2 - yCoord**2))
            else:
                xCoord = math.sqrt(max(0,sphere_radius**2 - zCoord**2 - yCoord**2))
            
            
            #Now I am going to rotate the X and Y coordinates counterclockwise using a rotation matrix
            xCoordTemp = xCoord
            xCoord = xCoord*math.cos(rho) - yCoord*math.sin(rho)
            yCoord = xCoordTemp*math.sin(rho) + yCoord*math.cos(rho)
            
            
            x_orig.append(xCoord)
            y_orig.append(yCoord)
            stereographCoord = stereographic(xCoord, yCoord, zCoord, sphere_radius)
            xStereo.append(stereographCoord[0])
            yStereo.append(stereographCoord[1])
            #print(phi, " ", zCoord, " ", xCoord, " ", stereographCoord[0], yCoord, " ", stereographCoord[1])
        #plt.plot(x_orig, y_orig)
        plt.plot(xStereo, yStereo, label=("theta: "+str(theta)+ " rho: " + str(rho)))
        maxCoord = max(maxCoord,abs(max(xStereo)),abs(max(yStereo)),abs(min(xStereo)),abs(min(yStereo)))
        minCoord = -maxCoord

minCoord = -4
maxCoord = 4
plt.ylim(minCoord, maxCoord)
plt.xlim(minCoord, maxCoord)
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.show()
            
```

<h6>a) Calculating Arc Distance to Equator for Randomly chosen points?</h6>

<br><p>A few observations stood out when I conducted this experiment for dimensions ranging from 2 to 1000. The first observation is that as the number of dimensions increased, that the variance of the distribution decreased proportionately, which is evidenced by the left tail moving further to the right. So for instance, if the number of dimensions increased from 10 to 100, then variance would decreasd by a factor of 10.

<br>What this says about uniform random points on a unit sphere is that given a large enough dimension size, that each point can be expected to be &pi;/2 radians away from the equator of the sphere, which is the maximum arc distance from the equator. Additionally I would expect this type of behavior to hold for any other type of <em>great circle</em>, where the arc distance between any given random uniform point and an arbitrary <em>great circle</em> would be &pi;/2 for large enough dimension size.</p>

```python
import numpy as np
import matplotlib.pyplot as plt
import math


sphereDimArr = [2, 4, 10, 100, 1000] #As given by the homework problem
for dim in sphereDimArr:
    sphereDim = dim
    N = sphereDim + 1
    equator = [0]*N
    equator[1] = 1 #any index can be set to a value of 1 except for index of 0, since we need equator[0] = 0.
    sampleSize = 10000
    distanceArr = [0]*sampleSize
    for i in range(0,sampleSize):
        x = np.random.normal(0, 1, size=N)
        x_norm = np.linalg.norm(x)
        sphere = x/x_norm
        dotProduct = np.dot(equator, sphere)
        angle = math.acos(dotProduct) #Note that a dot b = \\a\\*\\b\\*cos(theta)
        distanceArr[i] = min(math.pi-angle, angle) #this is because the angle can vary between 0 and pi,
                                          #but we're only interested in the angle between 0 and pi/2
    variance = np.var(distanceArr)
    print("Number of dimensions:", dim)
    print("Variance: ", variance)
    plt.style.use('ggplot')
    binNumber = int(sampleSize**0.5)
    plt.xlabel('Arc Distance From Equator (Radians)')
    plt.ylabel('Count')
    plt.hist(distanceArr, bins = binNumber) #Using the square root of number of sampled points.
    plt.show()
```

<h6>b) Calculating the angle between pairs of randomly generated points</h6>

<br><p>A few observations stood out when conducting this experiment for dimension sizes ranging from 2 to 1000. The first observation is that as the number of dimensions increased, that the variance of each distribution decreased proportionately to the magnitude of the dimension increase. So for instance, if the number of dimensions increased from 10 to 100, then variance would decreasd by a factor of 10.

<br>What we can conclude from this behavior is that given a large enough dimension size, that uniformly randomly generated points will be approximately at an angle of &pi;/2 with respect to each other. </p>

```python
import numpy as np
import matplotlib.pyplot as plt
import math


sphereDimArr = [2, 4, 10, 100, 1000] #As given by the homework problem
for dim in sphereDimArr:
    sphereDim = dim
    N = sphereDim + 1
    sampleSize = 1000
    angleArr = [0]*sampleSize
    for i in range(0,sampleSize):
        x1 = np.random.normal(0, 1, size=N)
        x1_norm = np.linalg.norm(x1)
        sphere_pt_1 = x1/x1_norm
        x2 = np.random.normal(0, 1, size=N)
        x2_norm = np.linalg.norm(x2)
        sphere_pt_2 = x2/x2_norm
        dotProduct = np.dot(sphere_pt_1, sphere_pt_2)
        angleArr[i] = math.acos(dotProduct) #Note that a dot b = \\a\\*\\b\\*cos(theta)

    
    variance = np.var(angleArr)
    print("Number of dimensions:", dim)
    print("Variance: ", variance)
    plt.style.use('ggplot')
    binNumber = int(sampleSize**0.5)
    plt.xlabel('Angle between pairs of random points')
    plt.ylabel('Count')
    plt.hist(angleArr, bins = binNumber) #Using the square root of number of sampled points.
    plt.show()
```
