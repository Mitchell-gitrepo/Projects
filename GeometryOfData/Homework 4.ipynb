<p>
$\gamma(t)=(a,be^{ct})$ is a geodesic if it's partial second derivative satisfies the below equation:<br><br>
$\frac{\partial^2 \gamma(t)^k}{\partial t^2} \space = \space -\sum\limits_{i,j=1}^{d} \Gamma_{i,j}^k (\gamma(t)) \frac{\partial \gamma^i}{\partial t}\frac{\partial \gamma^j}{\partial t}$<br><br>

Note that $\frac{\partial \gamma(t)}{\partial t} \space = \space [0, cbe^{ct}]^T$ and that $\frac{\partial^2 \gamma(t)}{\partial t^2} \space = \space [0,c^2be^{ct}]^T$<br><br>

Then let us consider the case of $k=1$. The only non-zero Christoffel symbols that contribute in this case are for $k=1,i=1,j=2$ and $k=1,i=2,j=1$ as shown above. Then: <br><br>
    
$\frac{\partial^2 \gamma(t)^1}{\partial t^2} \space = \space -(\Gamma_{1,2}^1 (\gamma(t)) \frac{\partial \gamma^1}{\partial t}\frac{\partial \gamma^2}{\partial t}+\Gamma_{2,1}^1 (\gamma(t)) \frac{\partial \gamma^2}{\partial t}\frac{\partial \gamma^1}{\partial t}) \space = \space -(\frac{-1}{\sigma(t)}(0)cbe^{ct} + \frac{-1}{\sigma(t)}cbe^{ct}(0))\space = \space 0$<br><br>

For the case of $k=2$ the only non-zero Christoffel symbols that contributes are for $k=2,i=1,j=1$ abd $k=2,i=2,j=2$. Then:

$\frac{\partial^2 \gamma(t)^2}{\partial t^2} \space = \space -(\Gamma_{1,1}^2 (\gamma(t)) \frac{\partial \gamma^1}{\partial t}\frac{\partial \gamma^1}{\partial t} + \Gamma_{2,2}^2 (\gamma(t)) \frac{\partial \gamma^2}{\partial t}\frac{\partial \gamma^2}{\partial t})\space = \space -(\frac{1}{2\sigma(t)}(0)(0) + \frac{-1}{\sigma(t)}cbe^{ct}*cbe^{ct})\space = \space -(\frac{-1}{be^{ct}}c^2b^2e^{2ct})\space = \space c^2be^{ct}$<br><br>
    
So then it has been shown that the proposed curve, $(µ(t), σ(t)) = (a, be^{ct})$, is a geodesic as it satisfies the above differential equation.

<p>The geodesic distance between the two gaussians with parameters (0,1) and (0,2) can be found by using the geodesic curve $(a, be^{ct})$. If $t=0$ corresponds with (0,1) and $t=1$ corresponds with (0,2), then:
<br><br> $(a, be^{c(0)}) = (0,1) \rightarrow a = 0; b = 1$
<br><br> $(a, be^{c(1)}) = (0,2) \rightarrow (0, 1e^{c(1)}) = (0,2) \rightarrow c = ln(2)$

<br><br> Then substituting these values into the length integral yields:
<br><br> $\int_{t=0}^{t=1} ||\gamma'(t)|| \,dt$ $\space$ = $\space$ $\int_{0}^{1} [0, cbe^{ct}]\begin{equation*}
\begin{pmatrix}
    \frac{1}{\sigma^2(t)} & 0 \\
    0 & \frac{2}{\sigma^2(t)}
    \end{pmatrix}
    \end{equation*}[0, cbe^{ct}]^T \,dt$ $\space$ = $\space$ $\int_{t=0}^{t=1} [0, \frac{2}{\sigma^2(t)}cbe^{ct}][0, cbe^{ct}]^T \,dt$ $\space$ = $\space$ $\int_{t=0}^{t=1} \frac{2}{\sigma^2(t)}c^2b^2e^{2ct} \,dt$ $\space$ <br><br> = $\space$ $\int_{t=0}^{t=1} \frac{2}{b^2e^{2ct}}c^2b^2e^{2ct} \,dt$ $\space$ = $\space$ $\int_{t=0}^{t=1} 2c^2 \,dt$ $\space$ = $\space$ $2c^2t|_0^1$ $\space$ = $\space$ $2(ln(2))^2(1) - 2(ln(2))^2(0)$  $\space$ = $\space$ $2(ln(2))^2$

<p>The geodesic distance between the two gaussians with parameters (-1,1) and (-1,1) can be found by using the geodesic curve $(atanh(ct)+b, a\frac{\sqrt{2}}{2}sech(ct))$. If $t=-1$ corresponds with (-1,1) and $t=1$ corresponds with (1,1), then:
<br><br> $atanh(-c)+b = -1$ and $atanh(c)+b = 1$, which means that $atanh(-c)+b + atanh(c)+b = -atanh(c)+b + atanh(c) + b = 2b = -1 + 1 = 0 \rightarrow b = 0$
<br><br>$atanh(c) = 1 \rightarrow a^2tanh^2(c) = 1$ and $a\frac{\sqrt{2}}{2}sech(c) = 1 \rightarrow a^2\frac{1}{2}sech^2(c) = 1 \rightarrow a^2tanh^2(c) + a^2sech^2(c) = 1 + 2 \rightarrow a^2(tanh^2(c) + sech^2(c)) = 3 \rightarrow a^2 = 3$ since $sech(c)$ is positive and $a\frac{\sqrt{2}}{2}sech(c) = 1$ then $a$ must be positive and equal to $\sqrt{3}$.
<br><br>$atanh(c) = 1 \rightarrow \sqrt{3}tanh(c) = 1 \rightarrow c = arctanh(\frac{\sqrt{3}}{3})$
</p>

The geodesic distance can be found using the length integral:
<br><br>$\int_{t=-1}^{t=1} ||\gamma'(t)|| \,dt$ $\space$ = $\space$ $\int_{-1}^{1} [ac(sech^2(ct)), -ca\frac{\sqrt{2}}{2}tanh(ct)sech(ct)]\begin{equation*}

    \end{equation*}[ac(sech^2(ct)), -ca\frac{\sqrt{2}}{2}tanh(ct)sech(ct)]^T \,dt$ $\space$ = $\space$ $\int_{t=-1}^{t=1} [\frac{acsech^2(ct)}{\sigma^2(t)},\frac{-\sqrt{2}ca(tanh(ct))sech(ct)}{\sigma^2(t)}][ac(sech^2(ct)), -ca\frac{\sqrt{2}}{2}tanh(ct)sech(ct)]^T \,dt$ $\space$ = $\space$ $\int_{t=-1}^{t=1} \frac{a^2c^2sech^4(ct)}{\sigma^2(t)} + \frac{c^2a^2tanh^2(ct)sech^2(ct)}{\sigma^2(t)} \,dt$ $\space$ <br><br> = $\space$ $\int_{t=-1}^{t=1} \frac{a^2c^2sech^4(ct)}{\frac{1}{2}a^2sech^2(ct)} + \frac{c^2a^2tanh^2(ct)sech^2(ct)}{\frac{1}{2}a^2sech^2(ct)} \,dt$ $\space$ = $\space$ $\int_{t=-1}^{t=1} 2c^2sech^2(ct) + 2c^2tanh^2(ct) \,dt$ $\space$ = $\space$ $\int_{t=-1}^{t=1} 2c^2 \,dt$ $\space$ = $\space$ $2c^2t|_{-1}^1$ $\space$ = $\space$ $2(arctanh(\frac{\sqrt{3}}{3}))^2(1) - 2(arctanh(\frac{\sqrt{3}}{3}))^2(-1)$  $\space$ = $\space$ $4(arctanh(\frac{\sqrt{3}}{3}))^2$

```python
from matplotlib import pyplot as plt
import numpy as np
import scipy.stats as stats
import math

#Normal Distribution Parameters
mu_1 = -1
mu_2 = 1
sigma_1 = 1
sigma_2 = 1

z1 = np.linspace(mu_1 - 6*sigma_1, mu_1 + 6*sigma_1, 100)
z2 = np.linspace(mu_2 - 3*sigma_2, mu_2 + 3*sigma_2, 100)

#Plotting Normal Distributions
plt.title("Gaussian Distributions with means of -1 and 1") 
plt.plot(z1, stats.norm.pdf(z1, mu_1, sigma_1))
plt.plot(z2, stats.norm.pdf(z2, mu_2, sigma_2))
plt.show()


#Parameters for Geodesic (See derivation in writeup)
a = math.sqrt(3)
b = 0
c = np.arctanh(1/math.sqrt(3))
x = []
y = []
for t in np.arange(-1, 1.01, 0.01):
    x_coordinate = a * np.tanh(c * t) + b
    y_coordinate = a * math.sqrt(2)/2 * 1 / np.cosh(c * t)
    x.append(x_coordinate)
    y.append(y_coordinate)

#plotting Poincare Half-Plane
plt.title("Gaussian Distributions in the Poincare Half-Plane") 
plt.plot(x, y)
plt.show()
print("done")
```

```python
from matplotlib import pyplot as plt
import numpy as np
import scipy.stats as stats
import math

#Normal Distribution Parameters
mu_1 = -1
mu_2 = 1
sigma_1 = 1
sigma_2 = 1

#Parameters for Geodesic
a = math.sqrt(3)
b = 0
c = np.arctanh(1/math.sqrt(3))
t = (mu_1 + mu_2) / 2
frechet_x = a * np.tanh(c * t) + b
frechet_y = a * math.sqrt(2)/2 * 1 / np.cosh(c * t)
print("The mu and sigma coordinates of the frechet mean are: ", "(",frechet_x,",",frechet_y,")")

z1 = np.linspace(mu_1 - 6*sigma_1, mu_1 + 6*sigma_1, 100)
z2 = np.linspace(mu_2 - 3*sigma_2, mu_2 + 3*sigma_2, 100)
z3 = np.linspace(frechet_x - 3*frechet_y, frechet_x + 3*frechet_y, 100)

#Plotting Normal Distributions
plt.title("Gaussian Distributions and Frechet Mean") 
plt.plot(z1, stats.norm.pdf(z1, mu_1, sigma_1))
plt.plot(z2, stats.norm.pdf(z2, mu_2, sigma_2))
plt.plot(z3, stats.norm.pdf(z3, frechet_x, frechet_y))
plt.show()
```

The $(\mu, \sigma)$ midpoint on the geodesic between them is (0, 1.2247). My main observation about the frechet mean is that it makes sense intuitively, in that if the mean of the two Gaussians are -1 and 1, then it makes intuitive sense that the Gaussian that represents the midpoint would have mean 0. Additionally this makes sense as a mean of 0 corresponds to the peak of the geodesic connecting the two Gaussians, which would be the point with the smallest distance since the distance between two points approaches $\infty$ as one of the points approaches the x-axis. 
<br><br> I think essentially the Frechet Mean gives a sense of what an "average" distribution looks like between two Gaussian distributions, and I would think that the level of dissimilarity, as given by the Kullback-Leibler divergence, between the Frechet Mean and one of the Gaussians would be the same for both Gaussians.

<br>Functions for Normal Gradient Descent

```python
import numpy as np
import math
from decimal import Decimal

#Formulas for implementation found here: https://www.textbook.ds100.org/ch/17/classification_cost.html

def initializeWeights(train_data):
    weights = -2 * torch.rand(train_data[0][0].shape) + 1
    weights = weights.flatten()
    weights = F.pad(weights, pad=(0, 1), mode='constant', value=0)
    return weights


def calc_sigma(weights, x):
    t = torch.dot(weights.flatten(), x.flatten()).item()
    
    #Occasionally I get an overflow error if t is of exceptionally large magnitude
    try:
        sigma = 1/(1+math.e**(-t))
    except:
        if t < 0:
            sigma = 0.0000000000001       #In case of overflow error
        else:
            sigma = 1
    return sigma


def calc_gradient(train_data, weights):
    n = len(train_data)
    sum = torch.zeros(weights.shape)
    for pt in train_data: 
        #values of y can only be 0 or 1 for this implementation
        if pt[1] == first_class:
            y = 1
        else:
            y = 0
        
        x = F.pad(pt[0].flatten(), pad=(0, 1), mode='constant', value=1)     #Extra dimension added to account for bias
        sum = sum + (y - calc_sigma(weights, x)) * x
    gradient = -1/n * sum
    
    return gradient


def gradientDescent(initial_weights, train_data, learning_rate, epoch):
    weights = initial_weights
    trainingLoss = []
    
    for i in range(0,epoch):
        curr_loss = calc_loss(train_data, weights)
        print("Current iteration loss: ", curr_loss)
        trainingLoss.append(curr_loss)
        grad = calc_gradient(train_data, weights)
        weights = weights - (learning_rate * grad)
        
    curr_loss = calc_loss(train_data, weights)
    trainingLoss.append(curr_loss)
    print("Current iteration loss: ", curr_loss)
    return weights, trainingLoss


def calc_loss(test_data, weights):
    n = len(test_data)
    sum = 0
    for pt in test_data:
        
        #values of y can only be 0 or 1 for this implementation
        if pt[1] == first_class:
            y = 1    
        else:
            y = 0
        x = F.pad(pt[0].flatten(), pad=(0, 1), mode='constant', value=1)     #Extra dimension added to account for bias
        wX = calc_sigma(weights, x)
        
        #In case of ln(0)
        try:
            sum = sum + (-y * math.log(wX) - (1 - y) * math.log(1 - wX))
        except:
            sum = sum + (-y * math.log(wX) - (1 - y) * math.log(1 - wX + 0.0000000000000001))
    loss = sum
    return loss 

```

<br>Functions for Natural Gradient Descent

```python
import numpy as np
import math


def log_liklihood_grad(train_data, weights):
    sum = torch.zeros(weights.shape)
    n = len(train_data)
    for pt in train_data:

        #values of y can only be 0 or 1 for this implementation
        if pt[1] == first_class:
            y = 1
        else:
            y = 0
        x = F.pad(pt[0].flatten(), pad=(0, 1), mode='constant', value=1)
        sum = sum + (y - calc_sigma(weights, x)) * x
    sum = sum / n
    return sum
    
    
def calc_fisher(train_data, weights):
    weights = weights.flatten()
    n = len(train_data)
    sum = torch.zeros(weights.shape[0], weights.shape[0])
    for pt in train_data:
        grad_log = log_liklihood_grad([pt], weights)
        sum = sum + torch.matmul(torch.transpose(grad_log.unsqueeze(0), 0, 1), grad_log.unsqueeze(0))
    sum = sum / n
    return sum


def calc_eta(train_sample, train_data, weights):
    fisher = calc_fisher(train_sample, weights)
    inverse_fisher = torch.pinverse(fisher)
    gradient = log_liklihood_grad(train_data, weights)
    eta = torch.matmul(inverse_fisher, gradient)
    return eta


def getRandSample(train_data, size):
    train_sample = []
    for i in range(0,len(train_data)):
        if (i%size) == 0:
            train_sample.append(train_data[i])
    return train_sample


def naturalGradientAscent(initial_weights, train_data, learning_rate, epoch):
    weights = initial_weights
    trainingLoss = []
    train_sample = getRandSample(train_data, 1) 
    for i in range(0,epoch):
        curr_loss = calc_loss(train_data, weights)
        print("Current iteration loss: ", curr_loss)
        trainingLoss.append(curr_loss)
        eta = calc_eta(train_sample, train_data, weights)
        weights = weights + (learning_rate * eta)
        
    curr_loss = calc_loss(train_data, weights)
    trainingLoss.append(curr_loss)
    print("Current iteration loss: ", curr_loss)
    return weights, trainingLoss


#The below function was coded but never actually used
def gradientAscent(train_data, learning_rate, epoch):
    weights = initializeWeights(train_data)
    trainingLoss = []
    
    for i in range(0,epoch):
        curr_loss = calc_loss(train_data, weights)
        print("Current iteration loss: ", curr_loss)
        trainingLoss.append(curr_loss)
        grad = log_liklihood_grad(train_data, weights)
        weights = weights + (learning_rate * grad)
        
    curr_loss = calc_loss(train_data, weights)
    trainingLoss.append(curr_loss)
    print("Current iteration loss: ", curr_loss)
    return weights, trainingLoss
```

<br>Five iterations of Normal Gradient Descent and Natural Gradient Descent for Logistic Predictor

```python
combined_weight_matrix = [[],[]]
combined_losses = [[],[]]
normal_learning_rate = .8      #Computed through trial and error
natural_learning_rate = .75    #Computed through trial and error - Note not perfect because initial loss varies a lot
epoch = 20

for i in range(0,5):
    initial_weight = initializeWeights(train_data)
    normal_weights, normal_TrainLoss = gradientDescent(initial_weight, train_data, normal_learning_rate, epoch)
    natural_weights, natural_TrainLoss = naturalGradientAscent(initial_weight, train_data, natural_learning_rate, epoch)
    
    combined_weight_matrix[0].append(normal_weights)
    combined_weight_matrix[1].append(normal_weights)
    combined_losses[0].append(normal_TrainLoss)
    combined_losses[1].append(natural_TrainLoss)
```
